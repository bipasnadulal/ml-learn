```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
```

```python
cancer_data = load_breast_cancer(as_frame=True)
cancer_df = cancer_data.data
cancer_df['target'] = cancer_data.target
```

```python
print(cancer_df.head(5))
print(cancer_df.shape)
```

```python
#counting missing values of each column and calculating the total sum.
print(cancer_df.isnull().sum())

#calculating the total values of target
print(cancer_df['target'].value_counts)
```

```python
#statistics summary (describe() method includes all, we can also calculate individually as .mean(), .median() etc)

print(cancer_df.describe())
```

```python
#creating the data ready for the model training
#splitting the dataset into train and test set

'''here X stores the features i.e. 
it stores all the columns except the one 'target' with .drop('column name', axis =) method'''
'''axis is used to indicates the pandas to operate on rows or columns
if axis = 0, operates on rows
axis = 1, operates on column
'''
X = cancer_df.drop('target', axis = 1) 

# y stores the label or the output column that we are trying to predict
y = cancer_df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 417) 
#setting a random_state to a specific number allows us to reproduce the same results every time we run the code. (it is simply setting the internal state to same point before generating or splitting)
```

```python
#instantiating the LinearSVC model with parameters: penalty, loss, C and random_state
model = LinearSVC(penalty="l2", loss="hinge", C=10, random_state=417)

#fitting/training the model
model.fit(X_train, y_train)
```

```python
test_accuracy = model.score(X_test, y_test)
print(test_accuracy)
```

```python
#Fine tuning the hyperparameters
fine_tuned_model = LinearSVC(penalty="l2", loss = "squared_hinge", C=10, max_iter=3500, random_state=417)
fine_tuned_model.fit(X_train, y_train)

test_accuracy = fine_tuned_model.score(X_test, y_test)
print(test_accuracy)
```
