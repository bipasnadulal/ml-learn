```python
# implementing the knn algorithm using euclidean distance method
import pandas as pd

banking_df = pd.read_csv("subscription_prediction.csv")

banking_df["y"] = banking_df["y"].apply(lambda x: 1 if x == "yes" else 0)

train_df = banking_df.sample(frac=0.85, random_state=417)
test_df = banking_df.drop(train_df.index)

X_train = train_df.drop("y", axis=1)
y_train = train_df["y"]

X_test = test_df.drop("y", axis=1)
y_test = test_df["y"]

def knn(feature, single_test_input, k):
    X_train['distance'] = abs(X_train[feature] - single_test_input)

    nearest_neighbors = X_train.nsmallest(k, 'distance')

    prediction = y_train.loc[nearest_neighbors.index].mode()[0]

    return prediction

model_prediction = knn(
    feature="age",
    single_test_input=X_test.iloc[108]["age"],
    k=3
)

print("Prediction:", model_prediction)

print("True Label:", y_test.iloc[108])


print("Correct Prediction?", model_prediction == y_test.iloc[108])
```

```python
# KNN - Multiple features

import pandas as pd

banking_df = pd.read_csv("subscription_prediction.csv")

banking_df["y"] = banking_df["y"].apply(lambda x: 1 if x=="yes" else 0)

banking_df_copy = banking_df.copy()
banking_df_copy = pd.get_dummies(data=banking_df_copy, columns=["marital"], drop_first=True)

train_df = banking_df_copy.sample(frac=0.85, random_state=417)
test_df = banking_df_copy.drop(train_df.index)

X_train = train_df.drop("y", axis=1)
y_train = train_df["y"]

X_test = test_df.drop("y", axis=1)
y_test = test_df["y"]

def knn(features, single_test_input, k):

    X_train["distance"] = 0

    for feature in features:

        X_train["distance"] += (
            X_train[feature].astype(int) -
            int(single_test_input[feature])
        ) ** 2

    prediction = y_train[
        X_train["distance"].nsmallest(n=k).index
    ].mode()[0]

    return prediction



prediction = knn(features =["age", "campaign", "marital_married", "marital_single"], single_test_input=X_test.iloc[108], k=3)

print(prediction)

predictions = []
for i in range(len(X_test)):
    pred = knn(
        features = ["age", "campaign", "marital_married", "marital_single"],
        single_test_input = X_test.iloc[i],
        k=3
    )
    predictions.append(pred)
    
X_test['predicted_y'] = predictions
correct_predictions = (X_test['predicted_y'] == y_test).sum()

accuracy = (correct_predictions / len(y_test)) * 100

print("Accuracy: ", accuracy)

```

```python
# normalizing features using min-max normalization / scaling

def knn(features, single_test_input, k):
    squared_distance = 0
    for feature in features:
        if X_train[feature].dtype == bool:
            train_val = X_train[feature].astype(int)
            test_val = int(single_test_input[feature])
            diff = train_val - test_val
        else:
            diff = X_train[feature] - single_test_input[feature]
        squared_distance += diff**2
        
    X_train["distance"] = squared_distance**0.5
    
    prediction = y_train[X_train["distance"].nsmallest(n=k).index].mode()[0]
    return prediction

age_min = X_train['age'].min()
age_max = X_train['age'].max()
X_train['age'] =(X_train['age'] - age_min)/(age_max - age_min)

X_test['age'] = (X_test['age'] - age_min) / (age_max - age_min)

campaign_min = X_train['campaign'].min()
campaign_max = X_train['campaign'].max()
X_train['campaign'] =(X_train['campaign'] - campaign_min)/(campaign_max - campaign_min)
X_test['campaign'] =(X_test['campaign'] - campaign_min)/(campaign_max - campaign_min)

predictions = []
for i in range(len(X_test)):
    pred = knn(
        features = ["age", "campaign", "marital_married", "marital_single"],
        single_test_input = X_test.iloc[i],
        k=3
    )
    predictions.append(pred)
X_test['predicted_y'] = predictions

correct_predictions = (X_test['predicted_y'] == y_test).sum()
model_accuracy = (correct_predictions / len(y_test) ) * 100

print("Accuracy: ", model_accuracy)
```
